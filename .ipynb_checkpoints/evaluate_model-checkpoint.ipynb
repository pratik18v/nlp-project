```python
import matplotlib.pyplot as plt
import cPickle as pickle
import tensorflow as tf
from my_core.solver import CaptioningSolver
from my_core.model import CaptionGenerator
from my_core.utils import load_coco_data
from my_core.bleu import evaluate

%matplotlib inline
plt.rcParams['figure.figsize'] = (8.0, 6.0)  # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

%load_ext autoreload
%autoreload 2
```

```python
data = load_coco_data(data_path='./data/data', split='val')
with open('./data/data/train/word_to_idx.pkl', 'rb') as f:
    word_to_idx = pickle.load(f)
```

```python
model = CaptionGenerator(word_to_idx, dim_att=27, dim_feature=[196, 512], dim_embed=512,
                                   dim_hidden=1024, n_time_step=16, prev2out=True, 
                                             ctx2out=True, alpha_c=1.0, selector=True, dropout=True)
```

```python
solver = CaptioningSolver(model, data, data, n_epochs=10, batch_size=64, update_rule='adam',
                                      learning_rate=0.001, print_every=50, save_every=10, image_path='./image/2014_resized',
                                pretrained_model=None, model_path='./model/lstm', test_model='./model/lstm1/model-10',
                                 print_bleu=False, log_path='./log/')
```

```python
solver.test(data, split='val')
```

```python
test = load_coco_data(data_path='./data/data', split='test')
```

```python
tf.get_variable_scope().reuse_variables()
solver.test(test, split='test')
```

```python
evaluate(data_path='./data', split='val')
```
